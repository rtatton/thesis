```python
def print_latex_table(_df: pl.DataFrame, network_type: str, parameter: str) -> None:
    table = (
        _df
        .filter(network_type=network_type)
        .select('nodes', 'edges', parameter)
        .sort('nodes', 'edges')
        .pivot('nodes', index='edges', values=parameter)
    )
    for row in table:
        print(' & '.join([row.name, *map(str, row.to_list())]) + ' \\\\')
```
